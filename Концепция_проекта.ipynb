# Основная идея.

Суть проекта заключается в создании системы, которая призвана помогать ГИБДД устанавливать причины аварий по остальным данным о ней. 

Такая система может пригодиться:
  * если причину аварии тяжело установить сразу (подсказка возможной причины аварии);
  * для уменьшения вероятности ошибки работника полиции при постановлении причины аварии;
  * и тд;


# Цели, задачи и прочее.

Реализовывать эту систему мы будем на основе данных об авариях путем машинного обучения. Мы, зная причины аварий в прошлом сможем выделить наиболее частые(то есть чтобы авария на эту причину была хотя бы не 1) и определять новые аварии к тому или иному классу.

## что обучаем?

Первостепенная задача научится предсказывать столбец violations, а конкретнее нарушения из этого столбца.

## План действий

У нас очень удачная команда для этого проекта, как я предлагаю распределить обязанности:
  1. (до 6 декабря) Тимофей, я и Леня начинаем выполнять п1 цз;
  2. (до 6 декабря) В это время Ваня и Тимур берутся за п3 цз;
  3. (до 11 декабря) После выполненных этих пунктов команда переходит к п2 цз:
    * Я и Тимофей беремся за первую звездочку; 
    * В это время Леня и Тимур выполняют врорую;
    * Ваня находится на связи, так как выбрал себе роль математика и экономиста, все такого рода вопросы в первую очередь летят к нему; 
  4. (с 11 декабря и до дедлайна) Как только выполнен пункт 2 цз мы с Тимофеем переходим к обучению и все оставшееся время тратим на него дабы сделать хорошо, остальная части команды в это время доделывает оставшиеся статистики, делает презентации, латает дыры, украшает графики и тд.
  5. Едем в Огниково и там добиваем все до идеала.

То, что пункт выполняется несколькими людьми подразумевает, что они сами договариваются о том, как будут выполнять поставленную задачу и сами решат что им делать, напомню, что важен не процесс работы, а результат.

# О комуникации и команде

Понятно, что нам не выжить если каждый будет делать что-либо у себя а потом как-то присылать результат, поэтому я создал гит-репозиторий, в который вас всех добавлю. При создании и изменении колабов, при изменении основного датасета, при желании прислать свои выкладки и отсчеты команде надо будет комитить в созданную для этого ветку репозитория. Для удобной работы с гитом я советую делать все комиты через пайчарм, если кто-либо из команды не умеет пользоваться гитом и не знает что это такое прошу сообщить, я могу вам организованно за пару минут объяснить основы и как что туда выложить.
